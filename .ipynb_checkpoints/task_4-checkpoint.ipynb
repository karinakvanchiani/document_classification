```python
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
```

```python
D = np.loadtxt('C:/Users/kvanc/session/document_classification/lin_reg.txt', delimiter=',')

X = torch.tensor(D[:, :-1], dtype=torch.float)
Y = torch.tensor(D[:, -1], dtype=torch.float)

X = torch.cat([torch.ones(X.shape[0], 1), X], dim=1)
```

1. Нарисуйте график среднеквадратичной ошибки в зависимости от параметра регуляризации $\alpha$, используя полные данные для обучения и для тестирования. 
2. Подготовьте исходные данные для 5 fold CV.
3. С помощью CV найдите оптимальное значение $\alpha$ на какой-нибудь сетке значений.
4. Постройте валидационную кривую. 
5. Постройте кривую обучения.

```python
mse = nn.MSELoss()

def get_weights(X, Y, alpha=1, learning_rate=0.001, batch_size=100, epochs=100):
    
    w = torch.rand(X.shape[1], requires_grad=True)
    losses = []
    old_w = 0
    num_batchs = X.shape[0] // batch_size
    
    for epoch in range(epochs):
        loss_acc = 0
        for i in range(num_batchs):
            X_batch = X[i * batch_size: (i + 1) * batch_size]
            y_batch = Y[i * batch_size: (i + 1) * batch_size]

            y_pred = torch.matmul(X_batch, w)

            loss = mse(y_pred, y_batch) + alpha * torch.norm(w)
            loss.backward()

            with torch.no_grad():

                grad = learning_rate * w.grad + 0.95 * old_w
                w -= grad
                old_w = grad
                w.grad.zero_()

                loss_acc += loss.item()

        losses.append(loss_acc / num_batchs)
    
    return losses, w
```

```python
def predict(X, weights):
    return torch.matmul(X, weights)
```

### mse

```python
alpha_losses = []
train_losses = []
space = np.arange(0.01, 10, 0.5)

for alpha in space:
    
    y_pred = predict(X, get_weights(X, Y, alpha=alpha, learning_rate=0.001)[1])
    alpha_losses.append(mse(Y, y_pred))
```

```python
plt.figure(figsize=(16, 9))

plt.plot(space, alpha_losses, label='mse')

plt.xlabel('alpha')
plt.ylabel('mse')

plt.legend()
plt.show()
```

### cross validation

```python
def indexes(shape, folds=5):
    
    ind = np.arange(shape)
    size = shape // folds

    for i in range(folds):
        train_indexes = ind[0: i * size]
        train_indexes = np.hstack((train_indexes, ind[(i + 1) * size:]))

        val_indexes = ind[i * size: (i + 1) * size]
        yield (train_indexes, val_indexes)
```

### search optimal alpha

```python
def cross_val_score(X, Y, alpha, folds=5):
    
    val_scores = []
    train_scores = []

    for (train, val) in indexes(X.shape[0], folds=folds):
        
        weights = get_weights(X[train], Y[train], alpha)[1]

        y_pred_train = predict(X[train], weights)
        train_scores.append(mse(Y[train], y_pred_train))
        
        y_pred_val = predict(X[val], weights)
        val_scores.append(mse(Y[val], y_pred_val))
        
    return torch.tensor(val_scores), torch.tensor(train_scores)
```

```python
val_losses = []
train_losses = []
space = np.arange(0.01, 10, 0.5)

for alpha in space:
    val_scores, train_scores = cross_val_score(X, Y, alpha=alpha)
    val_losses.append(torch.mean(val_scores))
    train_losses.append(torch.mean(train_scores))
```

```python
plt.figure(figsize=(18, 9))

plt.plot(space, val_losses, label='val mse')
plt.plot(space, train_losses, label='train mse')

plt.xlabel('alpha')
plt.ylabel('mse')

plt.legend()
plt.show()
```
